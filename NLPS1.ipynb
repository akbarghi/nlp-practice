```python
import nltk
```

```python
text="""Hello Mr. Smith, how are you doing today? The weather is great, and city is awesome.
... The sky is pinkish-blue. You shouldn't eat cardboard.
... Impulse-radio ultrawideband (IR-UWB) radar is a popular research topic in the field of post-earthquake search and rescue. 
... By lowering center frequency, it can penetrate through earthquake rubble to detect trapped victims mainly by identifying their respiratory-motion response. 
... Thus, low-center-frequency IR-UWB respiratory-motion response is characterized for the first time in this paper. 
... On this basis, a novel constant false alarm rate (CFAR) algorithm that automatically identifies the response is developed. 
... The IR-UWB respiratory-motion response has range extension and interrelation characteristics. With these characteristics, the algorithm can effectively improve the estimation accuracy of clutter energy in CFAR detection. 
... The characteristics and the algorithm performance are verified by experiment results, which show not only great promise in practice but also significance for future research of IR-UWB radar for detection of trapped victims."""

text
```

```python
#tokenize paragraf jadi sentences

from nltk.tokenize import sent_tokenize
tokenized_text=sent_tokenize(text)
tokenized_text
```

```python
#tokenize paragraph jadi words

from nltk.tokenize import word_tokenize
tokenized_word=word_tokenize(text)
print(tokenized_word)
```

```python
stopper = ['.','!','?']

sword = []
temp = []
for a in tokenized_word :
    if (a in stopper):
        sword.append(temp)
        temp = []
    else:
        temp.append(a)

print(sword)
```

```python
from nltk.util import bigrams, trigrams, ngrams
print(list(nltk.bigrams(tokenized_word)))
print('\n\n')
print(list(nltk.trigrams(tokenized_word)))
print('\n\n')
print(list(nltk.ngrams(tokenized_word, 5)))
#keanya ga gini dah

```

```python
from nltk.probability import FreqDist
fdist = FreqDist(tokenized_word)
print(fdist)
print(fdist.most_common(2))
fdist
```

```python
import matplotlib.pyplot as plt
fdist.plot(30, cumulative=False)
plt.show()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
ina_stop_words = set(stopwords.words("indonesian"))
print(stop_words)
print(ina_stop_words)
```

```python
filtered_sent = []
for w in tokenized_word:
    if w not in stop_words:
        filtered_sent.append(w)
        
print(filtered_sent)
```

```python
stopper = ['.','!','?']

filtered_sent2 = []
temp = ''
for a in tokenized_word:
    if a in stopper:
        temp = (str(temp)+' '+str(a))
        filtered_sent2.append(temp)
        temp = ''
    elif a not in stop_words:
        if temp == '':
            temp = (str(temp)+str(a))
        else:
            temp = (str(temp)+' '+str(a))
            #print(temp)
        
print(filtered_sent2)
```

```python
tokenized_pos_word = nltk.pos_tag(tokenized_word)
filtered_sent3 = []
tag = ["NN", "NNP", "VBP", "VBZ", "VBG", "JJ"]

for a in tokenized_pos_word:
    if a[1] in ["NN", "NNP", "VBP", "VBZ", "VBG", "JJ"]:
        filtered_sent3.append(a[0])
        
print(nltk.pos_tag(tokenized_word))
print('\n')
print(filtered_sent3)
```
